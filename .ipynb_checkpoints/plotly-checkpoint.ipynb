```python
import plotly.graph_objects as go

X = [0,1,2,3]
Y = [0,1,0,1]

data = go.Scatter(x=X, y=Y)
layout = {
    'title': "Zigzak with plotly"
}
fig = go.Figure(data, layout)
fig.show()
```

```python
salaries = [
    ("Mark", 1000),
    ("John", 1500),
    ("Daniel", 2300),
    ("Greg", 5000)]

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
data_visual = go.Bar(x=names, y=salary_values)
layout = {
    'title' : 'Salaries with plotly'
}
pict = go.Figure(data_visual, layout)
pict.show()
```
