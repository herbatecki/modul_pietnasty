```python
salaries = [1000, 5000, 7000]
salary_iterator = iter(salaries)
print(next(salary_iterator))
print(next(salary_iterator))
print(next(salary_iterator))
```

```python
# print(next(salary_iterator)) # gdy nie ma rzeczy do iterowania to wywala błąd
salary_iterator = iter(salaries)
salary_sum = 0
for salary in salary_iterator:
  salary_sum = salary_sum + salary
"Total salaries = %d" % salary_sum
```

```python
def give_me_first_ten_numbers():
  for i in range(1, 11):
    yield i

for number in give_me_first_ten_numbers():
  print(number)
```

```python
def sum_of_N_integers(n=100):
  my_number = 1
  while my_number<=n:
    my_number = my_number + 1
    yield my_number
```

```python
for sum in sum_of_N_integers():
  print(sum)
```

```python
import pandas as pd
pd.__version__
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]
```

```python
df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index('month')
df
```

```python
df['priceUSD'] = df.apply(lambda row: row.price / 4.0, axis=1)
print(df)
```

```python
import matplotlib.pyplot as plt
df['priceUSD'].plot(y=range(2), kind="line",linestyle="dashed", color="red", grid=True, ylim=(0,2))
```
