```python
names = ["Mark", "John", "Daniel", "Greg"]
salaries = [1000, 1500, 2300, 5000]
salary_series_improved = pd.Series(salaries, index=names) # przypisanie indeksów w postaci listy 'names'
salary_series_improved
```

```python
salary_series_raise = pd.Series([1100, 1600, 2400, 5100])
salary_raise = 100

for salary in salary_series_raise:
    print(salary+salary_raise)
```
