# Jak tworzyć funkcję w Pythonie?
używamy słowa "def" , następnie nazwę funkcji w raz z zamkniętymi nawiasami okrągłymi (miejsce na ewentualne parametry), dajemy dwukropek, enter, od następnej linii z wcięciem piszemy mechanizm naszej funkcji

```python
def sum_two_numbers(x, y):
    return x + y
```

do wywołanie funkcji używamy nazwy naszej funkcji, domnkniętych nawiasów okrągłych, a w nich w naszym przypadku nasze dwa parametry - dwie liczby:

```python
sum_two_numbers(4,6)
```
